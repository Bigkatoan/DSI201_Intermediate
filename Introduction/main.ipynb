## Data Science

Data science is an interdisciplinary field that involves extracting insights and knowledge from structured and unstructured data. It combines various techniques from statistics, mathematics, computer science, and domain knowledge to analyze complex datasets and derive actionable insights.

## Data Analyst, Data Engineer, Data Scientist

#### Data Analyst:

Role: Data analysts are primarily responsible for interpreting data, analyzing results, and providing insights to support decision-making. They work with structured data to identify trends, patterns, and correlations that can inform business strategies.

Skills: Strong analytical skills, proficiency in data analysis tools such as SQL, Excel, or statistical software like R or Python, data visualization skills using tools like Tableau or Power BI, and domain knowledge in the specific industry.

Tasks: Data cleaning, exploratory data analysis, creating reports and dashboards, performing ad-hoc analysis, and communicating findings to stakeholders.

#### Data Engineer:

Role: Data engineers focus on designing, building, and maintaining the infrastructure required for data generation, storage, and processing. They are responsible for developing data pipelines, ensuring data quality, and optimizing data workflows.

Skills: Proficiency in programming languages like Python, Java, or Scala, experience with big data technologies such as Hadoop, Spark, or Kafka, knowledge of database systems like SQL and NoSQL, and understanding of cloud platforms like AWS, Azure, or Google Cloud.

Tasks: Building and maintaining data pipelines, integrating data from multiple sources, ensuring data security and compliance, optimizing database performance, and collaborating with data scientists and analysts to meet their data requirements.

#### Data Scientist:

Role: Data scientists leverage statistical analysis, machine learning, and data mining techniques to extract insights from data and build predictive models. They apply advanced algorithms to solve complex problems and drive business value through data-driven decisions.

Skills: Strong background in statistics and mathematics, proficiency in programming languages like Python or R, expertise in machine learning algorithms and techniques, data visualization skills, and domain knowledge in the specific industry.

Tasks: Data cleaning and preprocessing, exploratory data analysis, feature engineering, model building and evaluation, deploying models into production, and communicating findings to non-technical stakeholders.

## Data.

In data science, "data" refers to raw facts, figures, or statistics collected for analysis. Data can be anything from numbers, words, measurements, observations, or descriptions of things. Data mainly split into 4 types of data:

- Nominal Data: Nominal data represents categories with no inherent order or ranking. Each value is distinct and unrelated to the others. Examples include gender (male, female), eye color (blue, brown, green), or types of cars (sedan, SUV, truck). In nominal data, you can't perform mathematical operations like addition or subtraction.
- Ordinal Data: Ordinal data also represents categories, but there's a clear order or ranking between them. However, the differences between the categories may not be uniform or quantifiable. Examples include survey responses with options like "poor," "fair," "good," and "excellent," or educational levels such as "high school," "college," "graduate school." In ordinal data, you can determine the order but not the magnitude of differences between categories.
- Discrete Data: Discrete data consists of separate, distinct values with no intermediate values possible within the given range. These values are typically integers. Examples include the number of children in a family (1, 2, 3, ...) or the number of cars passing through a toll booth in an hour. Discrete data is countable and finite within a particular range.
- Continuous Data: Continuous data represents measurements along a continuous scale and can take any value within a given range. It can include decimals and fractions. Examples include height, weight, temperature, and time. Continuous data can theoretically be measured with infinite precision, though in practice, there may be limitations due to measurement tools or practical constraints.

In 11 features, we can see there are salary, salary_currency, and salary_in_usd. We can reduce these feature.

```python
data = data.drop(['salary', 'salary_currency'], axis=1)
data.head()
```

Right now, I am pretty consider every jobs data science like in US.

```python
#preprocess data.
data = data[data['company_location'] == 'US']
data = data.sort_values('work_year')
```

What opportunities can I see :D

```python
years = data['work_year'].values
a, b = numpy.unique(years, return_counts=True)
plt.xlabel('years')
plt.ylabel('number of jobs')
plt.bar(a, b)
plt.show()
```

```python
salaries = data['salary_in_usd'].values
hash_map = {}
for val in set(years):
    hash_map.update({val: 0})

for i in range(len(years)):
    year = years[i]
    hash_map[year] += salaries[i]

nums = dict(zip(a, b))
for key in hash_map:
    hash_map[key] /= nums[key]
    
x, y = list(hash_map.keys()), list(hash_map.values())
plt.xlabel('years')
plt.ylabel('average salary in usd')
plt.bar(x, y)
plt.show()
```
